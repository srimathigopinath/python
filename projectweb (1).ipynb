```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://www.tickertape.in/blog/best-real-estate-stocks-in-india/"
```

```python
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.text)
soup.prettify()
```

```python
table = soup.find('table', {"class":"wp-block-table"})
table
```

```python
table=soup.find('tbody')
table
```

```python
rows=table.find_all("tr")
rows
```

```python
column_info=[]
for i in rows:
    cols=i.find_all("td")
    #print(cols)
    company_info=[c.text for c in cols]
    #print(country_info)
    column_info.append(company_info)
print(column_info)
```

```python
Name=[]
Market_cap=[]
Close_price=[]
PE_ratio=[]
Ayear_return=[]
```

```python
for i in column_info:
    Name.append(i[0])
    Market_cap.append(i[1])
    Close_price.append(i[2])
    PE_ratio.append(i[3])
    Ayear_return.append(i[4])
```

```python
Data={'Name:':Name,'Market_cap:':Market_cap,'Close_price:':Close_price,'PE_ratio:':PE_ratio,'Ayear_return:':Ayear_return}    
print(Data)
```

```python
realestate_info=pd.DataFrame(Data)
realestate_info.head(10)
```

```python
realestate_info.to_csv('Realestate_stocks.csv',index=False)
```
